```python
import drawsvg as draw
import math
```

```python
r = 12
row_width = 12
grid = []
n = 9
k = 1
for i in range(2, r+2):
    if n < i * math.pi / k:
        n *= 2
    cells = []
    for j in range(n):
        cells.append((i, j))
    grid.append(cells)
```

```python
w = h = 400
x = y = 200
d = draw.Drawing(w, h)
wall = draw.Group(
    fill='none',
    stroke='black',
    stroke_width=2
)

for row in grid:
    n = len(row)
    for r, a in row:
        radius = row_width * r
        arc_lenght = 360/n
        azimuth_1 = a * arc_lenght
        azimuth_2 = azimuth_1 + arc_lenght
        arc = draw.Arc(x, y, radius, azimuth_1, azimuth_2)
        wall.append(arc)

        x1 = x + radius * math.cos(math.radians(azimuth_2))
        y1 = y + radius * math.sin(math.radians(azimuth_2))
        x2 = x + (radius - row_width) * math.cos(math.radians(azimuth_2))
        y2 = y + (radius - row_width) * math.sin(math.radians(azimuth_2))
        line = draw.Lines(x1, y1, x2, y2)

        wall.append(line)
d.append(wall)
d
```
